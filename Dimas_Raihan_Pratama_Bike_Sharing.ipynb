### Cleaning Data

####**Menghapus tabel yang tidak diperlukan**

Sebelum masuk ke tahap selanjutnya, hapus tabel hr terlebih dahulu karena tidak relevan dengan pertanyaan yang sudah diterapkan serta nilai dari tabel hr dan day sama.

```python
del hr_df
```

#### **Menghapus beberapa kolom yang tidak digunakan**
1. instant --> tidak memberikan informasi yang dibutuhkan
2. windspeed --> tidak ada pertanyaan yang terkait kecepatan angin terhadap jumlah penyewa sepeda

```python
drop_col = ['instant', 'windspeed']

for i in day_df.columns:
  if i in drop_col:
    day_df.drop(labels=i, axis=1, inplace=True)

day_df.head()
```

####**Mengubah beberapa detail tentang kolom**

```python
# Mengubah nama judul kolom
day_df.rename(columns={
    'dteday': 'dateday',
    'yr': 'year',
    'mnth': 'month',
    'weathersit': 'weather_cond',
    'cnt': 'count'
}, inplace=True)

day_df.head()
```

```python
# Mengubah angka menjadi keterangan
day_df['month'] = day_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
day_df['weather_cond'] = day_df['weather_cond'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})
```

```python
# Mengubah tipe data ke datetime
day_df['dateday'] = pd.to_datetime(day_df.dateday)

# Mengubaha tipe data ke categorical
day_df['season'] = day_df.season.astype('category')
day_df['year'] = day_df.year.astype('category')
day_df['month'] = day_df.month.astype('category')
day_df['holiday'] = day_df.holiday.astype('category')
day_df['weekday'] = day_df.weekday.astype('category')
day_df['workingday'] = day_df.workingday.astype('category')
day_df['weather_cond'] = day_df.weather_cond.astype('category')

day_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: **Apakah terdapat peran cuaca terhadap jumlah pengguna sepeda?**

***Jawaban***: Dari visualisasi menggunakan boxplot, dapat dilihat bahwa terdapat hubungan antara kondisi cuaca dan jumlah penyewa sepeda. Kondisi cuaca saat sedang cerah atau setengah mendung menjadi kondisi yang diminati pengguna untuk menyewa sepeda. Kemudian, diikuti dengan cuaca berkabut atau mendung serta diikuti dengan cuaca sedikit bersalju atau hujan.

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x='weather_cond',
    y='count',
    data=day_df)

plt.title('Jumlah Pengguna Sepeda berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

### Pertanyaan 2: **Bagaimana kondisi ketika sepeda digunakan pada jam kerja, hari libur, dan hari kerja?**

***Jawaban***: Dari visualisasi menggunakan barplot, dapat dilihat bahwa:

1. Pengguna lebih banyak menyewa sepeda pada hari kerja daripada akhir pekan atau hari libur.
2. Pengguna lebih banyak menyewa sepeda lebih banyak pada hari biasa daripada hari libur(tanggal merah).
3. Pengguna paling banyak menyewa sepeda pada hari jumat dan sedikit pada hari minggu.

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(15,10))

# Berdasarkan workingday
sns.barplot(
    x='workingday',
    y='count',
    data=day_df,
    ax=axes[0])
axes[0].set_title('Jumlah Pengguna Sepeda berdasarkan Hari Kerja')
axes[0].set_xlabel('Hari Kerja')
axes[0].set_ylabel('Jumlah Pengguna Sepeda')

# Berdasarkan holiday
sns.barplot(
  x='holiday',
  y='count',
  data=day_df,
  ax=axes[1])
axes[1].set_title('Jumlah Pengguna Sepeda berdasarkan Hari Libur')
axes[1].set_xlabel('Hari Libur')
axes[1].set_ylabel('Jumlah Pengguna Sepeda')

# Berdasarkan weekday
sns.barplot(
  x='weekday',
  y='count',
  data=day_df,
  ax=axes[2])
axes[2].set_title('Jumlah Pengguna Sepeda berdasarkan Hari dalam Seminggu')
axes[2].set_xlabel('Hari dalam Seminggu')
axes[2].set_ylabel('Jumlah Pengguna Sepeda')

plt.tight_layout()
plt.show()
```

### Pertanyaan 3: **Seperti apa tren pengguna sepeda dalam satu tahun, Apakah tahun 2011 atau 2012 yang lebih unggul?**

***Jawaban***: Dari visualisasi menggunakan lineplot, dapat dilihat bahwa tahun 2012 lebih unggul dibandingkan 2011 dengan penyewaan tertinggi pada bulan September 2012 serta terendah pada bulan Januari 2012. Kemudian pada 2011 penyewaan tertinggi pada Juni 2011 dan terendah pada Januari 2011.

```python
day_df['month'] = pd.Categorical(day_df['month'], categories=
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    ordered=True)

monthly_counts = day_df.groupby(by=["month","year"]).agg({
    "count": "sum"
}).reset_index()

sns.lineplot(
    data=monthly_counts,
    x="month",
    y="count",
    hue="year",
    palette="rocket",
    marker="o")

plt.title("Jumlah total sepeda yang disewakan berdasarkan Bulan dan tahun")
plt.xlabel(None)
plt.ylabel(None)
plt.legend(title="Tahun", loc="upper right")
plt.tight_layout()
plt.show()
```

### Pertanyaan 4: **Seperti apa variabel temp dan atemp serta humidity dapat mempengaruhi jumlah total pengguna sepeda?**

***Jawaban***: Dari visualisasi menggunakan scatterplot, dapat dilihat bahwa:
1. Hubungan antara temp dan count menunjukkan korelasi yang positif.
2. Hubungan antara atemp dan count menunjukkan korelasi yang positif.
3. Hubungan antara hum dan count menunjukkan korelasi sedikit negatif tapi tidak signifikan karena tidak terlalu condong ke kiri.

Jadi, jika temperatur meningkat, maka pengguna yang menyewa sepeda cenderung meningkat dengan pengaruh dari humidity sangat kecil walaupun dapat menurunkan jumlah penyewa sepeda.

```python
plt.figure(figsize=(14, 6))

# Scatter plot untuk 'temp' vs 'count'
plt.subplot(1, 3, 1)
sns.scatterplot(
    x='temp',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Temperature vs Count')

# Scatter plot untuk 'atemp' vs 'count'
plt.subplot(1, 3, 2)
sns.scatterplot(
    x='atemp',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Feels Like Temperature vs Count')

# Scatter plot untuk 'hum' vs 'count'
plt.subplot(1, 3, 3)
sns.scatterplot(
    x='hum',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Humidity vs Count')
```

### Pertanyaan 5: **Apakah terdapat pengaruh musim terhadap penyewa sepeda berdasarkan pelanggan casual dan registered?**

***Jawaban***: Dari visualisasi menggunakan barplot, dapat dilihat bahwa dari semua musim, pengguna paling senang untuk menyewa sepeda pada musim gugur. Kemudian diikuti dengan musim panas, dingin dan semi.

```python
seasonal_usage = day_df.groupby('season')[['registered', 'casual']].sum().reset_index()

plt.figure(figsize=(10, 6))

plt.bar(
    seasonal_usage['season'],
    seasonal_usage['registered'],
    label='Registered',
    color='tab:blue'
)

plt.bar(
    seasonal_usage['season'],
    seasonal_usage['casual'],
    label='Casual',
    color='tab:orange'
)

plt.xlabel(None)
plt.ylabel(None)
plt.title('Jumlah penyewaan sepeda berdasarkan musim')
plt.legend()
plt.show()
```
